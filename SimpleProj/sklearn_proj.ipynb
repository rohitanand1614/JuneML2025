```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

```

```python
data_root = "https://github.com/ageron/data/raw/main/"
lifesdata = pd.read_csv(data_root + "lifesat/lifesat.csv")
```

```python
X = lifesdata[["GDP per capita (USD)"]].values
y = lifesdata["Life satisfaction"].values
```

```python
lifesdata.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500,62_500,4,9])
plt.show()
```

```python
#Select a linear model
model = LinearRegression()
```

```python
# Train the model
model.fit(X,y)
```

```python
X_new = [[37_655.2]]
print(model.predict(X_new))
```
